```python
cdf = (
    df['trending_date']
    .apply(lambda x: '20' + x[:2])  # Extract the year
    .value_counts()  # Count occurrences of each year
    .to_frame()  # Convert to DataFrame
    .reset_index()  # Reset index to make 'index' a column
    .rename(columns={'index': "year", 0: 'no_of_videos'})  # Rename columns
)
```

```python
print(cdf)
```

```python
# Ensure 'trending_date' column exists and contains valid data
if 'trending_date' in df.columns:
    # Extract the year part (e.g., '20YY') and count occurrences
    cdf = (
        df['trending_date']
        .apply(lambda x: '20' + x[:2])  # Extract the year as '20YY'
        .value_counts()  # Count occurrences of each year
        .to_frame()  # Convert Series to DataFrame
        .reset_index()  # Reset index to make 'index' a column
        .rename(columns={'index': 'year', 'trending_date': 'no_of_videos'})  # Rename columns
    )
    print(cdf.head())
else:
    print("Column 'trending_date' not found in DataFrame.")
```

```python
cdf.rename(columns={'no_of_videos': 'year', 'count': 'no_of_videos'}, inplace=True)

```

```python
#plotting the graph
plt.bar(cdf['year'], cdf['no_of_videos'])
plt.xlabel('Year')
plt.ylabel('Number of Videos')
plt.title('Trending Videos by Year')
plt.show()
```
